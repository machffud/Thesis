# Filterword  

filterword is representation of the article that consist of sentence, such that each sentence contain at least one keyword

```python
keyword_bencana = {
    'angin_topan':['badai', 'puting','beliung', 'topan', 'tornado', 'angin kencang'],
    'banjir':['banjir'],
    'erupsi':['vulkanik', 'erupsi', 'letusan', 'awan panas', 'lava'],
    'gempa':['gempa', 'tektonik'],
    'karhutla':['kebakaran hutan', 'kebakaran lahan', 'titik panas'],
    'kekeringan':['kekeringan'],
    'longsor':['longsor'],
    'tsunami':['tsunami']
}
def get_filter_word(bencana, keyword_bencana, df):
    col_with_window = []
    col_without_window = []
    for i in range(len(df)):
        tc = df['title'].loc[i]+'. '+df['text'].loc[i]
        cek_ada_keyword = False
        daftar_kalimat = tc.split('. ')
        indeks_window = [False for j in range(len(daftar_kalimat))]
        indeks_tanpa_window = [False for j in range(len(daftar_kalimat))]
        
        for indeks in range(len(daftar_kalimat)):
            kalimat = daftar_kalimat[indeks]
            for kata_kunci in keyword_bencana[bencana]:
                if kata_kunci in kalimat:
                    cek_ada_keyword = True
                    indeks_window[indeks] = True
                    indeks_tanpa_window[indeks] = True
                    
                    #additional for window
                    if indeks!=0:
                        indeks_window[indeks-1] = True
                    if indeks<len(daftar_kalimat)-1:
                        indeks_window[indeks+1] = True
                        
        result_with_window = ""
        result_without_window = ""
        for indeks in range(len(daftar_kalimat)):
            kalimat = daftar_kalimat[indeks]
            
            if indeks_window[indeks]:
                result_with_window+= kalimat+". "
            if indeks_tanpa_window[indeks]:
                result_without_window+= kalimat+". "
                
        if not cek_ada_keyword:
            result_with_window = df['title'].loc[i]
            result_without_window = df['title'].loc[i]
        
        col_with_window.append(result_with_window)
        col_without_window.append(result_without_window)
    df['filter_word_tc'] = col_without_window
    df['filter_word_tc_window'] =col_with_window
    return df,col_without_window
                        
#coba ,col_without_window = get_filter_word('banjir', keyword_bencana, df_asli.head(1))            
```

```python
def get_truncate(df):
    lst_first_short = []#5
    lst_last_short = []
    lst_first_long = []#10
    lst_last_long = []
    lst_combination = []
    
    df['tc_ori'] = df['title_ori']+'. '+df['text_ori']
    for i in range(len(df)):
        first_short, last_short, first_long, last_long, comb = '', '', '', '', ''
        daftar_kalimat = (df['title_ori'].loc[i]+'. '+df['text_ori'].loc[i]).split('. ')
        if len(daftar_kalimat)>=10:
            first_short = ". ".join(daftar_kalimat[:5])
            last_short = ". ".join(daftar_kalimat[-5:])
            first_long = ". ".join(daftar_kalimat[:10])
            last_long = ". ".join(daftar_kalimat[-10:])
            comb = ". ".join(daftar_kalimat[:5])+'. '+". ".join(daftar_kalimat[-5:])
        elif len(daftar_kalimat)>=5:
            first_short = ". ".join(daftar_kalimat[:5])
            last_short = ". ".join(daftar_kalimat[-5:])
            first_long = ". ".join(daftar_kalimat)
            last_long = ". ".join(daftar_kalimat)
            comb = ". ".join(daftar_kalimat)
        else:
            first_short = ". ".join(daftar_kalimat)
            last_short = ". ".join(daftar_kalimat)
            first_long = ". ".join(daftar_kalimat)
            last_long = ". ".join(daftar_kalimat)
            comb = ". ".join(daftar_kalimat)
        
        lst_first_short.append(first_short)#5
        lst_last_short.append(last_short)
        lst_first_long.append(first_long)
        lst_last_long.append(last_long)
        lst_combination.append(comb)
    return lst_first_short, lst_last_short,lst_first_long,lst_last_long,lst_combination
```
